```python
df_final_demo = pd.read_csv("df_final_demo.csv")

df_final_demo['clnt_tenure_yr'] = df_final_demo.clnt_tenure_yr.fillna(df_final_demo.clnt_tenure_yr.mean())
df_final_demo['clnt_tenure_mnth'] = df_final_demo.clnt_tenure_mnth.fillna(df_final_demo.clnt_tenure_mnth.mean())
df_final_demo['clnt_age'] = df_final_demo.clnt_age.fillna(df_final_demo.clnt_age.mean())

df_final_demo['gendr'] = df_final_demo['gendr'].fillna('U')
df_final_demo['num_accts'] = df_final_demo.num_accts.fillna(df_final_demo.num_accts.mean())
df_final_demo['bal'] = df_final_demo.bal.fillna(df_final_demo.bal.mean())
df_final_demo['calls_6_mnth'] = df_final_demo.calls_6_mnth.fillna(df_final_demo.calls_6_mnth.mean())
df_final_demo['logons_6_mnth'] = df_final_demo.logons_6_mnth.fillna(df_final_demo.logons_6_mnth.mean())

df_final_demo.head(50)
```

```python
concatenated_df = pd.concat([df_final_pt_1, df_final_pt_2], ignore_index=True)
concatenated_df.head(50)
```

```python
# Gets count of how many people were in each step of the process
confirm_count = concatenated_df['process_step'].value_counts()
print(confirm_count)
```

```python
# Gets avg duration (in sec) spent on each step
concatenated_df['date_time'] = pd.to_datetime(concatenated_df['date_time'])
concatenated_df = concatenated_df.sort_values(by=['client_id', 'date_time'])
concatenated_df['time_diff'] = concatenated_df.groupby('client_id')['date_time'].diff()
concatenated_df = concatenated_df.dropna(subset=['time_diff'])

avg_duration_per_step = concatenated_df.groupby('process_step')['time_diff'].mean()
avg_duration_per_step = avg_duration_per_step.apply(lambda x: x.total_seconds())

print(avg_duration_per_step)
```
