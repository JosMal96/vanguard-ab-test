#### avg_completion_time per group

```python
merged_df['client_id'] = merged_df['client_id'].astype(int)

# Filter out clients who reached the confirm step
clients_with_confirm = merged_df[merged_df['process_step'] == 'confirm']['client_id'].unique()
filtered_clients_df = merged_df[merged_df['client_id'].isin(clients_with_confirm)]

# Calculate the time difference between 'start' and 'confirm' for each client
start_times = filtered_clients_df[filtered_clients_df['process_step'] == 'start'].groupby('client_id')['date_time'].first()
confirm_times = filtered_clients_df[filtered_clients_df['process_step'] == 'confirm'].groupby('client_id')['date_time'].first()

completion_times = (confirm_times - start_times).reset_index()
completion_times.columns = ['client_id', 'completion_time']
completion_times['completion_time'] = completion_times['completion_time'].dt.total_seconds() / 60  # Convert to minutes

# Merge completion times with variation data
completion_times = completion_times.merge(merged_df, on='client_id')

completion_times = completion_times.dropna(subset=['completion_time'])
# completion_times.head(50)

average_completion_time = completion_times.groupby('variation')['completion_time'].mean()

print(average_completion_time)
```

### Hypothesis Testing

```python
# Check if the new design leads to a better user experience and higher process completion rates.

# Since the new design (Test group) had a higher completion rate compared to the old design (Control group), required to confirm if this difference is statistically significant.

# H0: The new design does not lead to a better user experience and higher process completion rates.
# H1: The new design does lead to a better user experience and higher process completion rates.
```

```python
# For the Hypothesis Testing:

# KPIs: Completion Rate, Time Spent on Each Step, Error Rates

# Check the variations (i.e. who was in which group - Control or Test)
# Check for the count of how many ppl came to the confirm step ~ Completion Rate
# Check for time spent on each step(i.e avg duration)
# Check for error rates(i.e. if user goes back to a previous step, it can indicate confusion or an error)
```

#### 1st Hypothesis

```python
# 1st hypothesis

# H0: completion time for test group == completion time for control group
# H1: completion time for test group != completion time for control group

# alpha = 0.05

# For this hypothesis, we can use just the p-value to determine
```

```python
# if the numbers are diff but the test is not showing much of a diff, then it means
# that the test is not very sensitive to the diff in the numbers(i.e. the diff is not statistically significant)
```

```python
alpha = 0.05
# completion_times = completion_times.merge(merged_df[['client_id', 'variation']].drop_duplicates(), on='client_id')
test_times = completion_times.loc[completion_times['variation'] == 'Test']['completion_time']

# completion_times.groupby('variation')['completion_time'].mean()

control_times = completion_times.loc[completion_times['variation'] == 'Control']['completion_time']

# completion_times.groupby('variation')['completion_time'].mean()
# 
t_stat, p_value = st.ttest_ind(test_times, control_times)
print(f"t-statistic: {t_stat}, p-value: {p_value}")

if p_value < alpha:
    print("Reject null hypothesis")
else:
    print("Fail to reject null hypothesis")
```

#### 2nd Hypothesis

```python
# 2nd Hypothesis

# H0: number of ppl who reached the confirm step in test group == number of ppl who reached the confirm step in the control group
# H1: number of ppl who reached the confirm step in test group != number of ppl who reached the confirm step in the control group

# alpha = 0.05
```

```python
# Z-Test for proportions
confirm_counts = step_counts['confirm']
total_test = step_counts.loc['Test'].sum()
total_control = step_counts.loc['Control'].sum()

print(f"Total Test Counts: {total_test}")
print(f"Total Control Counts: {total_control}")

# Number of confirms in each group
confirm_test = confirm_counts['Test']
confirm_control = confirm_counts['Control']

print(f"Confirm Test Counts: {confirm_test}")
print(f"Confirm Control Counts: {confirm_control}")
```

```python
# Z-test for Proportions
prop_test = confirm_test / total_test
prop_control = confirm_control / total_control

# Standard error
se = np.sqrt((prop_test * (1 - prop_test) / total_test) + (prop_control * (1 - prop_control) / total_control))

# Z-score
z = (prop_test - prop_control) / se

# P-value from Z-score
p_value = st.norm.sf(abs(z)) * 2  # two-tailed test

# Display results
print(f"Z-score: {z}")
print(f"P-value: {p_value}")

# Hypothesis testing
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```

#### 3rd Hypothesis

```python
# 3rd Hypothesis - can test if gender differences affect engaging with the new or old process

# H0: Gender distribution for test group == gender distribution for control group
# H1: Gender distribution for test group != gender distribution for control group

# alpha  = 0.05
```

```python
gender_df = pd.merge(df_final_demo, df_final_experiment, on='client_id')
gender_df = gender_df[['client_id', 'gendr', 'variation']]
gender_df.rename(columns={'gendr': 'gender'}, inplace=True)
gender_df

contingency_table = pd.crosstab(gender_df['variation'], gender_df['gender'])
print(contingency_table)

chi2_stat, p_value, dof, ex = st.chi2_contingency(contingency_table)
print(f"Chi-square statistic: {chi2_stat}, p-value: {p_value}")

alpha = 0.05

if p_value < alpha:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```
